```
def file_read_from_head(fname, nlines):
        from itertools import islice
        with open(fname) as f:
                for line in islice(f, nlines):
                        print(line)
file_read_from_head('Python.txt',2)
```

**WAP to read the first n lines of a file**

```
file = open("Python.txt","r")
x = file.readlines()
num = int(input("Enter num of lines to be read"))
for i,y in enumerate(x):
    if(i+1<=num):
        print(y, end=" ")
    
file.close()
```

**WAP to print the sum of the content of a file**

```
file = open("INT.txt","w")
file.close()
```

```
a = open('INT.txt', 'r') 
content = a.readlines() 
s = 0
for line in content:
  for i in line:
    if(i.isdigit() == True):
      s += int(i) 

print("The sum is:", s) 
```

**WAP to print the longest word in the whole line**

```
def longest_words(filename):
    with open(filename, 'r') as infile:
        words = infile.read().split()
    max_len = len(max(words, key=len))
    return [word for word in words if len(word) == max_len]
 
print(longest_words('Longword.txt'))
```

**WAP to print the content of a file in two diff o/p files**

```
file = open('abc.txt','w')
file.close()
```

```
with open('abc.txt', 'r') as f:
   lines = f.readlines()
even_lines, odd_lines = lines[::2], lines[1::2]

with open('abc1.txt','w') as f1:
   for odd_line in odd_lines:
       f1.write(odd_line)

with open('abc2.txt','w') as f2:
   for even_line in even_lines:
     f2.write(even_line)
```
